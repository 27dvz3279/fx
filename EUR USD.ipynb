```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
plt.style.use('default')
sns.set()
sns.set_style('darkgrid')
sns.set_palette('bright')

x = np.array([1, 3, 5, 7, 9])
gene_1 = np.array([1.397, 1.397, 1.397, 1.397, 1.397])
gene_2 = np.array([1.215, 1.215, 1.215, 1.215, 1.215])
gene_3 = np.array([1.033, 1.033, 1.033, 1.033, 1.033])
gene_4 = np.array([1.218, 1.233, 1.182, 1.168, 1.166])

plt.plot(x, gene_1, label='top')
plt.plot(x, gene_2, label='avg')
plt.plot(x, gene_3, label='bottom')
plt.plot(x, gene_4, label='EUR USD')

plt.legend()
plt.xlabel("month [2018]")
plt.ylabel("value")
plt.ylim(1.0, 1.45)
plt.show()
```
