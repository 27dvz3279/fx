```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
plt.style.use('default')
sns.set()
sns.set_style('darkgrid')
sns.set_palette('bright')

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
gene_1 = np.array([32000, 37000, 42700, 38000, 47000, 59000, 69000])
gene_2 = np.array([23200, 27200, 28800, 26000, 33000, 40000, 50700])
gene_3 = np.array([22800, 22800, 22800, 22800, 22800, 22800, 22800])
gene_4 = np.array([26000, 26000, 26000, 26000, 26000, 26000, 26000])
gene_5 = np.array([19000, 19000, 19000, 19000, 19000, 19000, 19000])

plt.plot(x, gene_1, label='SPX500')
plt.plot(x, gene_2, label='Nikkei')
plt.plot(x, gene_3, label='avg')
plt.plot(x, gene_4, label='top')
plt.plot(x, gene_5, label='bottom')

plt.legend()
plt.xlabel("time [year]")
plt.ylabel("value")
plt.ylim(15000, 80000)
plt.show()
```
