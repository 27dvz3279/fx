```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
plt.style.use('default')
sns.set()
sns.set_style('darkgrid')
sns.set_palette('bright')

x = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018])
gene_1 = np.array([12000, 14000, 18000, 20600, 20200, 22500, 26900])
gene_2 = np.array([8500, 10600, 16000, 17400, 18600, 19000, 23500])
gene_3 = np.array([16200, 16200, 16200, 16200, 16200, 16200, 16200])
gene_4 = np.array([18700, 18700, 18700, 18700, 18700, 18700, 18700])
gene_5 = np.array([13800, 13800, 13800, 13800, 13800, 13800, 13800])

plt.plot(x, gene_1, label='SPX500')
plt.plot(x, gene_2, label='Nikkei')
plt.plot(x, gene_3, label='avg')
plt.plot(x, gene_4, label='top')
plt.plot(x, gene_5, label='bottom')

plt.legend()
plt.xlabel("time [year]")
plt.ylabel("value")
plt.ylim(5000, 30000)
plt.show()
```
