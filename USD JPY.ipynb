```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
plt.style.use('default')
sns.set()
sns.set_style('darkgrid')
sns.set_palette('bright')

x = np.array([1, 2, 3, 4])
gene_1 = np.array([134.42, 134.42, 134.42, 134.42])
gene_2 = np.array([116.89, 116.89, 116.89, 116.89])
gene_3 = np.array([99.357, 99.357, 99.357, 99.357])
gene_4 = np.array([156.71, 155.08, 158.63, 159.25])

plt.plot(x, gene_1, label='top')
plt.plot(x, gene_2, label='avg')
plt.plot(x, gene_3, label='bottom')
plt.plot(x, gene_4, label='USD JPY')

plt.legend()
plt.xlabel("2026")
plt.ylabel("value")
plt.ylim(80, 180)
plt.show()
```
