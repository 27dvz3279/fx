```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
plt.style.use('default')
sns.set()
sns.set_style('darkgrid')
sns.set_palette('bright')

x = np.array([1, 3, 5, 7, 9])
gene_1 = np.array([120, 120, 120, 120, 120])
gene_2 = np.array([104.5, 104.5, 104.5, 104.5, 104.5])
gene_3 = np.array([89, 89, 89, 89, 89])
gene_4 = np.array([110.97, 106.03, 109.69, 111.47, 111.95])

plt.plot(x, gene_1, label='top')
plt.plot(x, gene_2, label='avg')
plt.plot(x, gene_3, label='bottom')
plt.plot(x, gene_4, label='USD JPY')

plt.legend()
plt.xlabel("month [2018]")
plt.ylabel("value")
plt.ylim(80, 130)
plt.show()
```
